# 1. library

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 2. Params

```python
os.chdir("D:/Projects/RoseTTAFold-All-Atom/")

dir_for_metrics_data = "./outputs/metrics of TCRs_MHCs.csv"

metrics_data = pd.read_csv(dir_for_metrics_data)
```

# 3. BoxPlot

```python
sns.set(style = "whitegrid")
plt.figure(figsize = (10, 6))

sns.boxplot(x = 'Protein_Type', y = 'mean_plddt', data = metrics_data, showfliers=False, color = 'white')
sns.stripplot(x = 'Protein_Type', y = 'mean_plddt', data = metrics_data, color = 'red', alpha = 0.5, jitter = True)

plt.title('Boxplot of mean_plddt for TCRs and MHCs', fontsize=14)
plt.xlabel('Protein_Type', fontsize=12)
plt.ylabel('mean_plddt', fontsize=12)


plt.tight_layout()
plt.savefig('./outputs/Boxplot_TCRs&MHCs.pdf', format='pdf')
plt.show()
```

```python
data = metrics_data[metrics_data['Protein_Type'] == "TCR"]

mean_plddt_mean = data['mean_plddt'].mean()
mean_plddt_std = data['mean_plddt'].std()

plt.figure(figsize=(10, 6))

plt.figure(figsize=(10, 6))
plt.scatter(data['Protein_Name'], data['mean_plddt'],
             alpha=0.7, color='b', marker='o')
plt.title('Scatter_Plot of mean_plddt  for TCR', fontsize=14)

plt.axhline(y=mean_plddt_mean, color='r', linestyle='--', label=f'Mean: {mean_plddt_mean:.2f}')
plt.axhline(y=mean_plddt_mean + mean_plddt_std, color='g', linestyle='-.', label=f'Mean + Std: {mean_plddt_mean + mean_plddt_std:.2f}')
plt.axhline(y=mean_plddt_mean - mean_plddt_std, color='g', linestyle='-.', label=f'Mean - Std: {mean_plddt_mean - mean_plddt_std:.2f}')


plt.xlabel('Protein_Name', fontsize=12)
plt.ylabel('mean_plddt', fontsize=12)
plt.legend(loc='lower left', bbox_to_anchor=(0, 0))

plt.tight_layout()

plt.savefig('./outputs/scatter_plot_TCR.pdf', format='pdf')
plt.show()
```
